```python
fib_cache = {}
def fibonacci(n):
    if n in fib_cache:
        return fib_cache[n]
    
    if n == 0:
        val = 1 # 0
    elif n == 1:
        val = 1    
    else:
        val = fibonacci(n-1) + fibonacci(n-2)
    
    fib_cache[n] = val
    return val

cap = 1000
for n in range(cap+1):
    # print(n, " : ", fibonacci(n))
    if n == cap:
        print(fibonacci(n)/fibonacci(n-1))
```

```python
i = 0
sums = 0
cap = 100
while i < cap+1 :
    sums = sums + i
    i = i + 1
print(sums)
print(sum(range(cap+1)))
```

```python
# Exercise 5.3
import pandas as pd

sigma = 10
beta = 8/3
rho = 28
dt = 0.01

x, y, z = 1, 1, 1

data = []
for time in range(2000):
    x, y, z = x + sigma*(y - x)*dt, y + (x*(rho - z) - y)*dt, z + (x*y - beta*z)*dt
    data.append([time, x, y, z])
    
df = pd.DataFrame(data, columns=["time", "x", "y", "z"])
df.plot.line(x="x", y="z", title="butterfly")
```
